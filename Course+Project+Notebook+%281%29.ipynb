```python
#def test_train_id(N, test_frac):
#    test_group = np.random.choice(np.arange(0,len(mouse_class)), size=(N*test_frac), replace=False, p=None)
#    train_group = np.setdiff1d(np.arange(0, len(mouse_class)), test_group)
#    return test_group, train_group
# 1080 = N
#test_train_id(1080, 0.3)
#test_group, train_group = test_train_id(1080, 0.3)
#print(test_group)
#print(train_group)

def test_train_id(N, test_frac):
    ntrain = np.floor(N * (1 - test_frac)).astype(int)
    randperm = np.random.permutation(N)
    train_group = randperm[:ntrain]
    test_group = randperm[(ntrain-N):]
    return train_group, test_group

```

```python
def cross_val_class_accuracy(model, X, y, r, test_frac, reps):
    proteins_pca = PCA()
    
    score = np.array([])
    for i in range(reps):
        train_group, test_group = test_train_id(y.size, test_frac)
        y_train = y[train_group]
        y_test = y[test_group]
        
        X_train = proteins_pca.fit_transform(X[train_group, :])[:, :r]
        X_test = proteins_pca.transform(X[test_group, :])[:, :r]
        
        model.fit(X_train, y_train)
        
        pred = model.predict(X_test)
        this_score = sum(pred == y_test) / len(y_test)
        
        score = np.append(score, this_score)
    return score
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis


score = cross_val_class_accuracy(LinearDiscriminantAnalysis(), processed_protein, 
                                mouse_class, 5, 0.2, 200)
print(score.mean())
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier

testfrac = 0.2
reps = 200

lda_r5_scores = cross_val_class_accuracy(LinearDiscriminantAnalysis(), 
                                          processed_protein, mouse_class, 5,
                                          testfrac, reps)
lda_r20_scores = cross_val_class_accuracy(LinearDiscriminantAnalysis(), 
                                          processed_protein, mouse_class, 16,
                                          testfrac, reps)
knn2_scores = cross_val_class_accuracy(KNeighborsClassifier(n_neighbors=2),
                                       processed_protein, mouse_class, 16,
                                       testfrac, reps)
knn10_scores = cross_val_class_accuracy(KNeighborsClassifier(n_neighbors=10),
                                       processed_protein, mouse_class, 16,
                                        testfrac, reps)
svm_scores = cross_val_class_accuracy(svm.SVC(kernel='linear'),
                                      processed_protein, mouse_class, 16, 
                                      testfrac, reps)
tree_scores = cross_val_class_accuracy(DecisionTreeClassifier(max_depth=3),
                                       processed_protein, mouse_class, 16,
                                       testfrac, reps)
```

```python
def cross_val_class_accuracy(model, X, y, r, test_frac, reps):
    cum_mcva = 0
    for i in range(reps):
        test_group, train_group = test_train_id(83, test_frac)
        traingroup_pca = PCA()
        traingroup_pca.fit(X[train_group, :])
        traingroup_reduced = traingroup_pca.transform(X[train_group, :])
    
        model.fit(traingroup_reduced[:, :r], y[train_group])
    
        testgroup_pca = PCA()
        testgroup_pca.fit(X[test_group, :])
        testgroup_reduced = testgroup_pca.transform(X[test_group, :])
    

        predictedvalues = model.predict(testgroup_reduced[:, :r])
        mcva = np.mean(y[test_group] == predictedvalues)
        cum_mcva = cum_mcva + mcva
    cvacc = cum_mcva/reps
    return cvacc
```
